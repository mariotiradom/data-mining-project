### DistribuciÃ³n del Nivel Educativo (Variable P301A)
Calculamos la frecuencia de cada nivel educativo y lo visualizamos en un grÃ¡fico de barras.

```python
# Contar frecuencias de la variable P301A
edu_counts = df_edu['P301A'].value_counts().sort_index()

# Crear el grÃ¡fico de barras horizontales
plt.figure(figsize=(10, 8))
sns.barplot(y=edu_counts.index.astype(int).astype(str), x=edu_counts.values, hue=edu_counts.index.astype(int).astype(str), palette='magma', legend=False)

plt.title('DistribuciÃ³n por Nivel Educativo Alcanzado (P301A)')
plt.ylabel('CÃ³digo de Nivel Educativo')
plt.xlabel('Cantidad de Personas')
plt.grid(axis='x', linestyle='--', alpha=0.6)

plt.show()

# Mostrar tabla de frecuencias para referencia
print("Frecuencias por nivel:")
print(edu_counts)
```
